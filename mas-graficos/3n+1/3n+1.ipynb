# El caso $ 3n + 1 $

```python
import matplotlib.pyplot as plt
```

### El algoritmo consiste en
* Si n es par (n/2)
* Si n es impar (3n + 1)

```python
def cuenta(lista, n):
    lista.append(n)

    if (n % 2) == 0:
        res = n / 2
    else:
        res = (3 * n) + 1

    if n != 1:
        cuenta(lista, res)
```

```python
res1 = []
res2 = []
res3 = []
res4 = []
numero1 = 85
numero2 = 77
numero3 = 67
numero4 = 46

cuenta(res1, numero1)
cuenta(res2, numero2)
cuenta(res3, numero3)
cuenta(res4, numero4)

cant_numeros1 = len(res1)
cant_numeros2 = len(res2)
cant_numeros3 = len(res3)
cant_numeros4 = len(res4)
```

```python
x = [x for x in range(1, cant_numeros1 + 1)]
y = res1

z = [x for x in range(1, cant_numeros2 + 1)]
j = res2

k = [x for x in range(1, cant_numeros3 + 1)]
l = res3

m = [x for x in range(1, cant_numeros4 + 1)]
n = res4
```

```python
plt.figure(figsize=(20, 9))

plt.stem(x, y, bottom=1, linefmt='k:', markerfmt='#6897BB')
plt.stem(z, j, bottom=1, linefmt='k:', markerfmt='#1f1f1f')
plt.stem(k, l, bottom=1, linefmt='k:', markerfmt='#BD4267')
plt.stem(m, n, bottom=1, linefmt='k:', markerfmt='#FF4000')
# o sino,
# plt.plot(x, y, '#6897BB', z, j, '#1f1f1f', k, l, '#BD4267', m, n, '#39B7AC')

plt.ylabel('resultado de la función según par o impar', size=15)
plt.xlabel('cantidad de operaciones realizadas', size=15)

plt.title("Secuencias", pad=35, loc="center",
          fontdict={'fontsize': 25, 'fontweight': 'bold', 'backgroundcolor': 'white'})
plt.show()
```
